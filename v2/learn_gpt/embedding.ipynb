```python
from gensim.models import Word2Vec
from gensim.test.utils import common_texts
import numpy as np

def l2_distance(vec1, vec2):
    distance = np.linalg.norm(vec1 - vec2)
    return distance

def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    similarity = dot_product / (norm_vec1 * norm_vec2)
    return similarity

print(common_texts)

# 训练模型
model = Word2Vec(sentences=common_texts, vector_size=128, window=5, min_count=1, workers=4)

# 查看单词 'computer' 的词向量
cp_vector = model.wv['computer']
m_vector = model.wv['human']
tr_vector = model.wv['trees']
print(len(cp_vector))
print("human and computer: ", cosine_similarity(cp_vector, m_vector), l2_distance(cp_vector, m_vector))
print("trees and computer: ", cosine_similarity(cp_vector, tr_vector), l2_distance(cp_vector, tr_vector))
print("human and trees: ", cosine_similarity(m_vector, tr_vector), l2_distance(m_vector, tr_vector))
```
